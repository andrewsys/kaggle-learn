```python
import pandas as pd
reviews = pd.read_csv('./datasets/winemag-data-130k-v2.csv', index_col=0)
pd.set_option('display.max_rows', 5) #Maybe not applicable in PyCharm
```

```python
# Getting data from the row '0'
reviews.iloc[0] # Same as reviews.iloc[0, :]
```

```python
# iloc and loc are row-first, columns-second.
# reviews.iloc[:, 0]
# reviews.iloc[:3, 0]
# reviews.iloc[5:10, 0]
# reviews.iloc[[0, 2, 5], 0]
reviews.iloc[-5:]
```

```python
# Label-based selection
reviews.loc[0, 'country']
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

```python
# iloc or loc?
# when using iloc the selection '0:10' excludes the last value, so the selected entry will be 0, ..., 9.
# meanwhile loc select indexes inclusively, so '0:10' selects 0, ..., 10.
# as you can see 0:10 iloc returns 10 values, but 0:10 loc returns 11.
```

```python
# indexes are no immutable, the set_index() method can easily manipulate them
reviews.set_index('title')
```

```python
reviews.loc[(reviews['country'] == 'Italy') & (reviews['points'] >= 90)]
```

```python
reviews.loc[(reviews['country'] == 'Italy') | (reviews['points'] >= 90)]
```
